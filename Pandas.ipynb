```python
df=pd.read_csv(r'/content/anime.csv')
print(df)
df.head()
df.tail()
df.shape
df.info
df.columns
df.describe
df.iloc[2]['Title']
```

```python
def extract_episodes(txt):
  data=""
  check=False
  for i in txt:
    if i==')':
      check=False
      return data
    if check:
      data+=i
    if i=='(':
      check=True
```

```python
df['Episodes']=df['Title'].apply(extract_episodes)
df['Episodes']=df['Episodes'].str.replace(' eps','')
df['Episodes']
df['Episodes']=pd.to_numeric(df['Episodes'])
df['Episodes'].mean()
df.index
```

```python
df.iloc[0]['Title']
```

```python
def extract_time(txt):
  data=""
  check=False
  count=0
  for i in txt:
    if count==19:
      return data
    if check:
      count+=1
      data+=i
    if i==")":
      check=True
print(extract_time(df.iloc[0]['Title']))
```

```python
df['Time']=df['Title'].apply(extract_time)
df['Time']

df['Score']==df['Score'].max()
df[df['Score']==df['Score'].max()]
print(df['Title'].value_counts().head())
df['Title'].head()
df[df['Episodes']==df['Episodes'].max()]
print(df['Episodes'].value_counts().head(5))
df.sort_values(by='Episodes',ascending=False).head(5)
```

```python
def extract_months(duration):
    start, end = [x.strip() for x in duration.split("-")]
    start = pd.to_datetime(start, format="%b %Y")
    end = pd.to_datetime(end, format="%b %Y")
    return (end.year - start.year) * 12 + (end.month - start.month)

```

```python
df['Months']=df['Time'].apply(extract_months)
df['Months']
df[df['Months']==df["Months"].max()]
```
